```python
import string
# Открытие файла и чтение его содержимого
with open('Уэллс_-_Война_миров.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    text.replace(',','')
    text.replace('.','')
    text.replace('','')
```

```python
words = text.split()
num_chars = len(text)
num_no_spaces = len(text.replace(' ', ''))
num_lines = text.count('\n') + 1
for i in words:
  if i == '–':
    words.pop(words.index(i))
num_words = len(words)
print(f"Количество символов (с пробелами): {num_chars}")
print(f"Количество символов (без пробелов): {num_no_spaces}")
print(f"Количество строк: {num_lines}")
print(f"Количество слов: {num_words}")
```

```python
# Удаление знаков препинания из текста
dell = str.maketrans('', '', string.punctuation)
text = text.translate(dell)
```

```python
word_count = {}
for word in words:
    if word.lower() in word_count:
        if word.lower() != '–':
          word_count[word.lower()] += 1
    else:
        word_count[word.lower()] = 1
word_count['было']
```

```python
import matplotlib.pyplot as plt
top_words = sorted(word_count, key=word_count.get, reverse=True)[:20]
counts = [word_count[word] for word in top_words]
plt.bar(range(len(top_words)), counts, align='center')
plt.xticks(range(len(top_words)), top_words, rotation=45)
plt.xlabel('Слова')
plt.ylabel('Количество')
plt.title('20 наиболее употребительных слов')
plt.show()
```
